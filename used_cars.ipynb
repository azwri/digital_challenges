```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/old_file_used_car_cleaned.csv")
df2 = pd.read_csv("data/old_file_UsedCarsSA_Clean_EN.csv")
```

```python
df1.head(1)
```

```python
df2.head(1)
```

```python
df1_copy = df1.copy()
```

```python
df1_copy.rename(columns={"car_brand":"brand", "car_model":"model", "car_driven":"driven", "car_transmission":"gear", "car_model_year":"year", "car_price":"price"}, inplace=True)
```

```python
df2_copy = df2.copy()
df2_copy.rename(columns={"Make":"brand", "Type":"model", "Mileage":"driven", "Gear_Type":"gear", "Year":"year", "Price":"price"}, inplace=True)
df2_copy = df2_copy.iloc[:, [0,1,9,8,2,11]]
```

```python
used_cars = pd.concat([df1_copy, df2_copy])
```

```python
used_cars.to_csv("data/used_cars.csv", index=False)
```
